Transformation alters the relationships between variables. In some cases, it may exaggerate or attenuate the relationships between variables.
For example, if there is a non-linear relationship between ApplicantIncome and another variable, square root transformation may distort or magnify this relationship.
Non-linear Transformations:
While square root transformation can reduce skewness, it imposes a specific non-linear structure on the data.
This transformation may not adequately capture the underlying relationships or distributional properties of the original data.
Interpretability:
Transformed variables may be less interpretable. For instance, interpreting the effect of a one-unit change in the square root of ApplicantIncome may be less intuitive compared to the original ApplicantIncome.
Bias in Predictive Models:
Square root transformation may introduce bias in predictive models if the relationship between the transformed feature and the target variable is not well-captured by the transformation.
This bias may lead to inaccuracies in model predictions, especially if the transformation disproportionately affects certain segments of the data.

## 5.6 SMOTE Technique üî®
üëâ In previous exploration, it can be seen that **the number between approved and rejected loan is imbalanced**. In this section, **oversampling technique will be used to avoid overfitting**,

#WRONG IMPLIMENTATION!

Using SMOTE on the entire dataset (X) is generally not recommended for several reasons:

Data Leakage:
SMOTE generates synthetic samples to balance the classes by interpolating between existing minority class samples. If SMOTE is applied to the entire dataset, information from the test set can leak into the training process, leading to overfitting and overly optimistic performance estimates.
Models trained on data with synthetic samples created from the entire dataset may not generalize well to unseen data.
Synthetic Samples in Test Set:
When SMOTE is applied to the entire dataset, the synthetic samples generated by SMOTE will be included in the test set. This can lead to biased evaluation results since the model has seen synthetic samples during training.
The test set should remain independent of the training process to accurately evaluate model performance.
Increased Complexity and Overfitting:
SMOTE may introduce unnecessary complexity to the majority class by generating synthetic samples. This can lead to overfitting and reduced model generalization.
SMOTE should be applied only to the minority class to balance the class distribution without introducing unnecessary complexity to the majority class.
Inflation of Class Ratios:
Applying SMOTE to the entire dataset may lead to an excessive inflation of class ratios, making the dataset unrepresentative of the true underlying distribution.
This can lead to skewed evaluations and misleading conclusions about model performance.

Yes, applying SMOTE to the X_train data can introduce bias with regard to fairness, especially if the original dataset already contains bias.

Here's how SMOTE may introduce bias in fairness:

Amplification of Existing Bias:
If the original dataset contains biases related to sensitive attributes (such as gender or race), applying SMOTE can potentially amplify these biases.
SMOTE generates synthetic samples by interpolating between existing minority class samples, which may inherit the biases present in the original data.
Impact on Fairness Metrics:
Fairness metrics, such as disparate impact or equal opportunity, may be affected by the introduction of synthetic samples through SMOTE.
If the synthetic samples created by SMOTE disproportionately affect certain groups, it can lead to biased outcomes in fairness evaluations.
Model Performance Bias:
Biases introduced by SMOTE can affect the performance of machine learning models, leading to disparities in predictions across different groups.
Models trained on biased data may produce unfair outcomes, favoring or disadvantaging certain groups in their predictions.

#WRONG IMPLIMENTATION!

When using MinMaxScaler to scale the features, it's important to fit the scaler on the training data (X_train) and then transform both the training and test data.

Here's why:

Prevent Data Leakage:
Fitting the scaler on the entire dataset (X) may lead to data leakage because information from the test set is used during the training process. This can result in overly optimistic performance estimates and poor generalization to unseen data.
Maintain Consistency:
Scaling should be performed consistently on both training and test datasets to ensure that the same scaling factors are applied to all features. This maintains the integrity of the data and ensures that the model learns from and generalizes well to unseen data.
Realistic Evaluation:
Scaling should reflect the distribution of the training data to provide a realistic representation of the data for model training. Scaling the test set separately could lead to different data distributions between the training and test sets, affecting model performance evaluation.

The False Positive Rate (FPR) measures the proportion of actual negative cases that are incorrectly classified as positive by the model. In the context of loan eligibility prediction, it represents the rate at which individuals who should not be granted a loan are mistakenly approved.
If the FPR is significantly higher for the unmarried group compared to the married group, it suggests that the model is more prone to incorrectly approving loans for unmarried individuals.

Higher FPR for Females: This means that the model is incorrectly classifying more female applicants as creditworthy (positive) when they are actually not, leading to a higher rate of false approvals or loans being granted to individuals who should not receive them.
In other words, a higher FPR for females suggests that the model is more likely to mistakenly approve loans for female applicants who may default, resulting in higher financial risks for the lending institution.
